# Limpieza de datos

```python
df = df.drop_duplicates()
```

```python
## eliminar columnas que no aportan valor analítico
columns_to_drop = ['Lat','Long']
df = df.drop(columns=columns_to_drop)   
```

# Gestión de Nulos

```python
# Rellenamos los nulos con etiquetas genéricas
df['Tweet Location'] = df['Tweet Location'].fillna('Desconocido')
df['Tweet Language'] = df['Tweet Language'].fillna('Desconocido')

## podria generado otras opciones


df['Verified or Non-Verified'] = df['Verified or Non-Verified'].fillna('Desconocido')
df['Name'] = df['Name'].fillna('Desconocido')   
```

```python
# 🛠️ CORRECCIÓN DE TIPOS NUMÉRICOS
df['Retweets Received'] = pd.to_numeric(df['Retweets Received'], errors='coerce')
df['Likes Received'] = pd.to_numeric(df['Likes Received'], errors='coerce')
df['User Followers'] = pd.to_numeric(df['User Followers'], errors='coerce').fillna(0).astype(int)
# Rellenamos los posibles valores corruptos (que ahora son NaN) con 0 interacciones
df['Retweets Received'] = df['Retweets Received'].fillna(0).astype(int)
df['Likes Received'] = df['Likes Received'].fillna(0).astype(int)
df['User Followers'] = df['User Followers'].fillna(0).astype(int)
```

```python
#La columna de fecha viene como texto. La convertimos a formato datetime real.
df['Tweet Posted Time (UTC)'] = pd.to_datetime(df['Tweet Posted Time (UTC)'], errors='coerce')
```

```python
df.info()
```

```python
# Creamos columnas derivadas para facilitar la segmentación en el Dashboard
df['Año-Mes'] = df['Tweet Posted Time (UTC)'].dt.to_period('M').astype(str)
df['Hora del Día'] = df['Tweet Posted Time (UTC)'].dt.hour
```

```python
df.head()
```

```python
# Aseguramos que los nombres de las categorías estén estandarizados
df['Verified or Non-Verified'] = df['Verified or Non-Verified'].str.strip()
```

```python
# la columna geográfica es uno de los mayores desafíos (y oportunidades) en esta prueba técnica.
# Si miramos la columna Tweet Location, verás que es un campo de texto libre. Eso significa que un usuario escribe "Madrid, Spain", otro "BCN", otro "En mi casa" y muchos lo dejan vacío

# El objetivo de este paso es crear una nueva columna limpia llamada Pais_Estandarizado mediante técnicas de búsqueda de palabras clave (Keywords).


# Rellenamos los nulos para poder aplicar funciones de texto sin que de error
df['Tweet Location'] = df['Tweet Location'].fillna('Desconocido').astype(str)

def extraer_pais(texto):
    texto_min = texto.lower() # Convertimos a minúsculas para facilitar la búsqueda
    
    # Mapeo de palabras clave a Países Estandarizados
    if 'spain' in texto_min or 'españa' in texto_min or 'madrid' in texto_min or 'barcelona' in texto_min:
        return 'España'
    elif 'usa' in texto_min or 'united states' in texto_min or 'estados unidos' in texto_min or 'california' in texto_min or 'ny' in texto_min:
        return 'Estados Unidos'
    elif 'uk' in texto_min or 'united kingdom' in texto_min or 'reino unido' in texto_min or 'london' in texto_min or 'londres' in texto_min:
        return 'Reino Unido'
    elif 'mexico' in texto_min or 'méxico' in texto_min or 'cdmx' in texto_min:
        return 'México'
    elif 'argentina' in texto_min or 'buenos aires' in texto_min:
        return 'Argentina'
    elif 'colombia' in texto_min or 'bogota' in texto_min or 'bogotá' in texto_min:
        return 'Colombia'
    elif 'italy' in texto_min or 'italia' in texto_min or 'roma' in texto_min:
        return 'Italia'
    elif 'france' in texto_min or 'francia' in texto_min or 'paris' in texto_min:
        return 'Francia'
    elif 'desconocido' in texto_min:
        return 'No Especificado'
    else:
        return 'Otros / No Identificado' # Para ubicaciones raras como "En mi nube" o países con pocos tweets

# 3. Aplicamos la función fila por fila creando la nueva columna
df['Pais_Estandarizado'] = df['Tweet Location'].apply(extraer_pais)

# 4. Revisamos rápidamente cómo quedó la distribución
print("--- Distribución de Tweets por País Estandarizado ---")
print(df['Pais_Estandarizado'].value_counts())
```

Agrupar a los usuarios en "Categorías de Influencia" según sus seguidores.

```python
#Nos aseguramos de que las columnas críticas sean estrictamente numéricas
df['User Followers'] = pd.to_numeric(df['User Followers'], errors='coerce').fillna(0).astype(int)
df['Retweets Received'] = pd.to_numeric(df['Retweets Received'], errors='coerce').fillna(0).astype(int)

def clasificar_influencia(seguidores):
    seg = int(seguidores)
    if seguidores < 500:
        return 'Micro-usuario (0-500)'
    elif seguidores < 5000:
        return 'Usuario Activo (501-5k)'
    elif seguidores < 50000:
        return 'Micro-Influencer (5k-50k)'
    else:
        return 'Celebridad / Institución (>50k)'

df['Categoria_Influencia'] = df['User Followers'].apply(clasificar_influencia)
```

En Twitter es crucial saber si el tuit es una opinión aislada o si intenta interactuar. Viendo el contenido de Tweet Content, muchos tuits incluyen links (https://...) o mencionan a la Organización Mundial de la Salud (@WHO) o a políticos (@Trump).

```python
# Verificamos si el tuit menciona a alguien (contiene '@')
df['Tiene_Mencion'] = df['Tweet Content'].str.contains('@', na=False).map({True: 'Sí', False: 'No'})

# Verificamos si comparte un enlace externo (contiene 'http')
df['Tiene_Enlace'] = df['Tweet Content'].str.contains('http', na=False).map({True: 'Sí', False: 'No'})
```

La columna Client nos dice desde dónde se publicó el tuit (Twitter for iPhone, Twitter Web App, TweetDeck, etc.). Hay cientos de variantes.

```python
df['Client'].unique()
```

```python
def agrupar_dispositivo(cliente):
    cliente_min = str(cliente).lower()
    if 'iphone' in cliente_min or 'ipad' in cliente_min:
        return 'Apple iOS'
    elif 'android' in cliente_min:
        return 'Android'
    elif 'web' in cliente_min:
        return 'Web (Escritorio)'
    elif 'tweetdeck' in cliente_min or 'hootsuite' in cliente_min:
        return 'Herramientas Profesionales'
    else:
        return 'Otros / Bots'

df['Dispositivo_Limpio'] = df['Client'].apply(agrupar_dispositivo)
```
